```
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```
df = pd.read_csv("/content/drive/MyDrive/Learning/Deep_Learning/Data.csv",encoding="ISO-8859-1")
```

```
train = df[df['Date']<='20150101']
test =  df[df['Date']>'20150101']
```

```
data_train = train.iloc[:,2:27]
data_train = data_train.replace("[^a-zA-z]"," ")
comment=[]
for i in range(data_train.shape[0]):
  comment.append(" ".join([str(dat) for dat in data_train.iloc[i,0:25]]))
cv = CountVectorizer(ngram_range=(2,2))
X_train= cv.fit_transform(comment).toarray()
y=train['Label']
```

```
data_test = test.iloc[:,2:27]
data_test = data_test.replace("[^a-zA-z]"," ")
comment=[]
for i in range(data_test.shape[0]):
  comment.append(" ".join([str(dat) for dat in data_test.iloc[i,0:25]]))
X_test= cv.transform(comment).toarray()
```

```
model = RandomForestClassifier()
model.fit(X_train,y)
y_pred = model.predict(X_test)
from sklearn.metrics import confusion_matrix
confusion_matrix(test['Label'],y_pred)
```
